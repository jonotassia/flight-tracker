```python
from amadeus_helpers import parse_date_range, generate_price_by_date_range, graph_price_by_date_range
import pandas as pd
from matplotlib import pyplot as plt
from collections import deque
```

```python
# Create queue for flight data
flight_queue = deque()
```

```python
# Define query parameters for SYD to LHR
origin = "SYD"
destination = "LHR"
num_adults = 2
departure_range = pd.date_range(start="2023-03-17",end="2023-03-24")
return_range = pd.date_range(start="2023-07-26", end="2023-07-30")

# Parse flights
flights_SYD_LHR_df = parse_date_range(origin, destination, departure_range, return_range, num_adults)

# Save to load in future if needed
flights_SYD_LHR_df.to_csv("./data/output/SYD-LHR.csv")

# Generate price dataframe
price_SYD_LHR_df = generate_price_by_date_range(flights_SYD_LHR_df)

# Append to price deque
flight_queue.append(price_SYD_LHR_df)
```
